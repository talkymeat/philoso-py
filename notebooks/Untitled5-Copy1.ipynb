```python
text_1 = """Bear in mind that bear puns on 'bear' are superior to bear puns on 'bare', which in turn are superior to bear puns on 'Bayer'"""

text_2 = """'Bare', after all, never refers to ursids, but only to nudity ('bare bottom'), emptyness ('a bare  cupboard'), or lack of adornment ('a bare wall'). Still, it can be used in more contexts than Bayer."""

text_3 = """Bayer is just a boring pharmaceuticals company, which some people pronounce as a disyllable anyway. Fuck Bayer"""

regex_bear = r'[bB]ear'
regex_bare = r'[bB]are'
regex_bayer = r'[bB]ayer'

regex_bear_context = r'.{0,8}[bB]ear.{0,8}'
regex_bare_context = r'.{0,8}[bB]are.{0,8}'
regex_bayer_context = r'.{0,8}[bB]ayer.{0,8}'

for text in [text_1, text_2, text_3]:
    print(re.findall(regex_bear_context, text))
    print(re.findall(regex_bare_context, text))
    print(re.findall(regex_bayer_context, text))
```

```python
def find_bears(search_text):
    bears = re.findall(r'.{0,8}[bB]ear.{0,8}', search_text)
    bares = re.findall(r'.{0,8}[bB]are.{0,8}', search_text)
    bayers = re.findall(r'.{0,8}[bB]ayer.{0,8}', search_text)
    return bears or bares or bayers
```

```python
print(find_bears(text_1))
```

```python
print(find_bears(text_2))
```

```python
print(find_bears(text_3))
```

```python
def replace_matches(s):
    p = re.compile(".ug")
    matches = [p.search(s)]
    i=0
    next_match = p.search(s, matches[i].end())
    while next_match:
        matches.append(next_match)
        i += 1
        next_match = p.search(s, matches[i].end())

    new_s = ""
    for m in matches[-1::-1]:
        new_s = s[m.start():].replace(m.group(0), '=-=') + new_s
        s = s[:m.start()]

    s += new_s
    return s
```
