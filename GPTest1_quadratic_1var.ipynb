```python
from gp import GP
import pandas as pd
import operators as ops
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.DataFrame({'x': [float(x) for x in range(-100, 101)]})
df['y'] = 3 * df['x']* df['x'] - 9* df['x'] + 18
df
```

```python
gp = GP(mutation_rate = 0.2, mutation_sd=1.0, operators=[ops.SUM, ops.PROD, ops.SQ, ops.POW, ops.CUBE])
```

```python
res = gp.run_gp(400, df['y'], 100, pd.DataFrame(df['x']), gp.create_seed_polynomial, 
                elitism=5, best_tree=True, rmses=True, best_rmse=True, 
                order=2, const_min=-20, const_max=20)
```
